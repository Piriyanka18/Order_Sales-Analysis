```python
import pandas as pd
df= pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
#rename the columns
#df.columns= df.columns.str.lower()
#df.columns
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

```python
#drive the new columns discount , sale price and profit
df['discount']=df['list_price']*df['discount_percent']*0.01
#sales price
df['sales_price']=df['list_price'] - df['discount']
# profit
df['profit'] = df['sales_price']-df['cost_price']
df.head()
```

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
#drop the columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```

```python
import sqlalchemy as sal
from sqlalchemy.engine import URL

# Define the connection parameters
server = "DESKTOP-DOJJ05F\\SQLEXPRESS"
database = "master"
driver = "ODBC+Driver+17+for+SQL+Server"

# Create the connection URL
connection_url = URL.create(
    "mssql+pyodbc",
    query={
        "odbc_connect": f"DRIVER={{{driver}}};SERVER={server};DATABASE={database};Trusted_Connection=yes;"
    }
)

# Create the engine
engine = sal.create_engine(connection_url)

# Connect to the database
try:
    conn = engine.connect()
    print("Connection successful!")
except Exception as e:
    print(f"Connection failed: {e}")

# Close the connection
#conn.close()
```
